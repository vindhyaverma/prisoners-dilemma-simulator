### Steps 3-8: Implementing the Simulator
I will now combine the logic for the payoff matrix, AI strategies, and the Gradio interface into the following cells.

```python
import random
import gradio as gr

def get_scores(move1, move2):
    if move1 == "Cooperate" and move2 == "Cooperate":
        return 3, 3
    elif move1 == "Betray" and move2 == "Cooperate":
        return 5, 0
    elif move1 == "Cooperate" and move2 == "Betray":
        return 0, 5
    else:
        return 1, 1

def ai_move(strategy, opponent_last_move):
    if strategy == "Always Cooperate":
        return "Cooperate"
    elif strategy == "Always Betray":
        return "Betray"
    elif strategy == "Tit-for-Tat":
        return opponent_last_move if opponent_last_move else "Cooperate"
    else: # Random
        return random.choice(["Cooperate", "Betray"])

def play_game(player1_strategy, player2_strategy, rounds):
    score1 = 0
    score2 = 0
    result = ""
    p1_last_move = None
    p2_last_move = None

    for i in range(int(rounds)):
        move1 = ai_move(player1_strategy, p2_last_move)
        move2 = ai_move(player2_strategy, p1_last_move)

        points1, points2 = get_scores(move1, move2)
        score1 += points1
        score2 += points2

        p1_last_move = move1
        p2_last_move = move2

        result += f"Round {i+1}\n"
        result += f"Player A ({move1}) vs Player B ({move2})\n"
        result += f"Round Score: {points1}-{points2} | Total: {score1}-{score2}\n\n"

    result += "--- FINAL RESULTS ---\n"
    result += f"Player A Total Score: {score1}\n"
    result += f"Player B Total Score: {score2}\n"

    if score1 > score2:
        result += "WINNER: Player A!"
    elif score2 > score1:
        result += "WINNER: Player B!"
    else:
        result += "RESULT: It's a Draw!"

    return result

# Create Gradio Interface
app = gr.Interface(
    fn=play_game,
    inputs=[
        gr.Dropdown(["Always Cooperate", "Always Betray", "Random", "Tit-for-Tat"], label="Player A Strategy"),
        gr.Dropdown(["Always Cooperate", "Always Betray", "Random", "Tit-for-Tat"], label="Player B Strategy"),
        gr.Slider(1, 20, value=5, step=1, label="Number of Rounds")
    ],
    outputs="text",
    title="Prisoner's Dilemma Strategy Simulator (Enhanced)",
    description="Compare strategies like Tit-for-Tat, Betrayal, and Cooperation."
)

app.launch()
```

### Explanation of Strategies

*   **Always Cooperate**: A pacifist strategy that always chooses to cooperate, regardless of what the opponent does. It is vulnerable to exploitation.
*   **Always Betray**: A purely selfish strategy that always chooses to betray. It maximizes short-term gain if the opponent cooperates but leads to poor outcomes if both players use it.
*   **Random**: A strategy with no memory or logic, choosing between cooperation and betrayal with a 50/50 chance.
*   **Tit-for-Tat**: A highly effective strategy in game theory. It starts with cooperation and then simply copies the opponent's previous move. It rewards cooperation and punishes betrayal immediately.
